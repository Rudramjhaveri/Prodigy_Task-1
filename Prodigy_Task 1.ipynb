```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/DELL/Downloads/Prodigy_Task 1/population.csv")
```

```python
data.head(266)
```

```python
# List of new countries for the graph
countries_to_plot = ['Australia', 'Austria', 'Argentina', 'Armenia', 'Andorra', 'Angola']

# Filtering the data frame
filtered_data = data[data['Country Name'].isin(countries_to_plot)]

# Creating the bar chart for the year 2010
plt.bar(filtered_data['Country Name'], filtered_data['2010'])
plt.xlabel('Country Name')
plt.ylabel('Population in 2010')
plt.title('Population of selected countries in 2010')
plt.xticks(rotation=50, ha='right')  # Rotating the x-axis labels for readability
plt.show()
```
